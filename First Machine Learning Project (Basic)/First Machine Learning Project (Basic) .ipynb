
 # ***MODEL Persisting***



````python
# Done..... Succesfully trained and Saved our Module ....Now No Need to do it always......
music_data=pd.read_csv('music.csv')
X=music_data.drop(columns=['genre']) # we are split data into Input(Gender and Age) and Output(Genre) HERE X IS INPUT
y=music_data['genre'] # y is Output


model = DecisionTreeClassifier()
model.fit(X, y)
joblib.dump(model, 'music-recommender.joblib')
````

````python
# now load when needed 
model = joblib.load('music-recommender.joblib')
predictions =  model.predict([[21,1]])
predictions
````
